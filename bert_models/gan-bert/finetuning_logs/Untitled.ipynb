```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gan_bert_50_res = np.load('gan_bert_finetuned_sst2_67349_samples_0.5_labelratio_2022-05-04_00-59-36.npy', allow_pickle=True)
gan_bert_20_res = np.load('gan_bert_finetuned_sst2_67349_samples_0.2_labelratio_2022-05-08_08-34-28.npy', allow_pickle=True)
gan_bert_70_res = np.load('gan_bert_finetuned_sst2_67349_samples_0.7_labelratio_2022-05-08_15-18-12.npy', allow_pickle=True)
```

```python
# Gan 20

gan_bert_20_train_loss_gen = [dic['Training Loss generator'] for dic in gan_bert_20_res]
gan_bert_20_train_loss_desc = [dic['Training Loss discriminator'] for dic in gan_bert_20_res]
gan_bert_20_val_loss =  [dic['Valid. Loss'] for dic in gan_bert_20_res]
gan_bert_20_val_accur =  [dic['Valid. Accur.'] for dic in gan_bert_20_res]
```

```python
# Gan 50

gan_bert_50_train_loss_gen = [dic['Training Loss generator'] for dic in gan_bert_50_res]
gan_bert_50_train_loss_desc = [dic['Training Loss discriminator'] for dic in gan_bert_50_res]
gan_bert_50_val_loss = [dic['Valid. Loss'] for dic in gan_bert_50_res]
gan_bert_50_val_accur = [dic['Valid. Accur.'] for dic in gan_bert_50_res]
```

```python
# Gan 70

gan_bert_70_train_loss_gen = [dic['Training Loss generator'] for dic in gan_bert_70_res]
gan_bert_70_train_loss_desc = [dic['Training Loss discriminator'] for dic in gan_bert_70_res]
gan_bert_70_val_loss =[dic['Valid. Loss'] for dic in gan_bert_70_res]
gan_bert_70_val_accur = [dic['Valid. Accur.'] for dic in gan_bert_70_res]
```

```python
 plt.figure(figsize=(12, 5))
    
plt.plot(gan_bert_20_train_loss_gen)
plt.plot(gan_bert_50_train_loss_gen)
plt.plot(gan_bert_70_train_loss_gen)
plt.ylabel('Generator loss')
plt.xlabel('Epochs')
plt.title('Generator loss during Training vs Epochs')
plt.legend(['0.2% labeled', '0.5% labeled', '0.7% labeled'])
plt.show()
```

```python
 plt.figure(figsize=(12, 5))
    
plt.plot(gan_bert_20_train_loss_desc)
plt.plot(gan_bert_50_train_loss_desc)
plt.plot(gan_bert_70_train_loss_desc)
plt.ylabel('Discriminator loss')
plt.xlabel('Epochs')
plt.title('Discriminator loss during Training vs Epochs')
plt.legend(['0.2% labeled', '0.5% labeled', '0.7% labeled'])
plt.show()
```

```python
 plt.figure(figsize=(12, 5))
    
plt.plot(gan_bert_20_val_loss)
plt.plot([i for i in gan_bert_50_val_loss])
plt.plot(gan_bert_70_val_loss)
plt.ylabel('Test loss')
plt.xlabel('Epochs')
plt.title('Test loss during Training vs Epochs')
plt.legend(['0.2% labeled', '0.5% labeled', '0.7% labeled'])
plt.show()
```

```python
 plt.figure(figsize=(12, 5))
    
plt.plot(gan_bert_20_val_accur)
plt.plot([i for i in gan_bert_50_val_accur])
plt.plot(gan_bert_70_val_accur)
plt.ylabel('Test Accuracy')
plt.xlabel('Epochs')
plt.title('Test accuracy during training vs Epochs')
plt.legend(['0.2% labeled', '0.5% labeled', '0.7% labeled'])
plt.show()
```
